```python
import requests
import pandas as pd
import json
import configparser
```

```python
config = configparser.ConfigParser()
config.read('/content/config.ini')
key = config['API']['key']
```

```python

def movies(key,query_title,year):

  movie_info_list = []

  for query_title,year in zip(query_title, year):
    r=requests.get(f'https://www.omdbapi.com/?apikey={key}&t={query_title}&y={year}')
    data = r.json()
    #print(data)

    movie_info_list.append(data)

  dfmovie = pd.DataFrame(movie_info_list)
  return dfmovie


#13 Filmes
query_title =[
    "Halloween",
    "Halloween II",
    "Halloween III: Season of the Witch",
    "Halloween 4: The Return of Michael Myers",
    "Halloween 5: The Revenge of Michael Myers",
    "Halloween: The Curse of Michael Myers",
    "Halloween H20: 20 Years Later",
    "Halloween: Resurrection",
    "Halloween ",
    "Halloween II",
    "Halloween ",
    "Halloween Kills",
    "Halloween Ends"
]
year = [
    "1978",
    "1981",
    "1982",
    "1988",
    "1989",
    "1995",
    "1998",
    "2002",
    "2007",
    "2009",
    "2018",
    "2021",
    "2022"
]
```

```python
#visualização do dataframe fora da função
movies(key,query_title,year)
```

```python
#transformação do Dataframe pra uso fora da função movies
dfmovies = movies(key, query_title, year)
```

```python
#Colunas desnecessárias
dfmovies.drop(['Rated'],axis=1,inplace=True)
dfmovies.drop(['Released'],axis=1,inplace=True)
dfmovies.drop(['Runtime'],axis=1,inplace=True)
dfmovies.drop(['Genre'],axis=1,inplace=True)
dfmovies.drop(['Director'],axis=1,inplace=True)
dfmovies.drop(['Writer'],axis=1,inplace=True)
dfmovies.drop(['Actors'],axis=1,inplace=True)
dfmovies.drop(['Plot'],axis=1,inplace=True)
dfmovies.drop(['Language'],axis=1,inplace=True)
dfmovies.drop(['Country'],axis=1,inplace=True)
dfmovies.drop(['Awards'],axis=1,inplace=True)
dfmovies.drop(['Poster'],axis=1,inplace=True)
dfmovies.drop(['Ratings'],axis=1,inplace=True)
dfmovies.drop(['Metascore'],axis=1,inplace=True)
dfmovies.drop(['Type'],axis=1,inplace=True)
dfmovies.drop(['DVD'],axis=1,inplace=True)
dfmovies.drop(['BoxOffice'],axis=1,inplace=True)
dfmovies.drop(['Production'],axis=1,inplace=True)
dfmovies.drop(['Website'],axis=1,inplace=True)
dfmovies.drop(['Response'],axis=1,inplace=True)
dfmovies.drop(['imdbID'],axis=1,inplace=True)
dfmovies.drop(['imdbVotes'],axis=1,inplace=True)
```
